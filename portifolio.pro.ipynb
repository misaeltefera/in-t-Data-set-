```python
# Sort by 'Revenue' in descending order
df1 = df.sort_values('Revenue', ascending=False)

print(df1)
```

```python
df1 = df[['Cost', 'Country']].sort_values(by='Cost', ascending=False).head(5)
print(df1)
```

```python
df1.plot.pie(y='Cost', labels=df1['Country'], autopct='%1.1f%%', legend=False)
```

```python
df2 = df.groupby('State')['Profit'].sum().sort_values(ascending=False)
df2
```

```python
# Plotting
df2.plot(kind='bar', x='State', y='Profit')
plt.show()
```
